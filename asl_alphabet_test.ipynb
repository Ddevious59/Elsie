```python
# -------------------------------
# 1. Install Libraries
# -------------------------------
# Uncomment these if running locally or in Colab for the first time
# !pip install mediapipe opencv-python tensorflow scikit-learn matplotlib seaborn

# -------------------------------
# 2. Import Libraries
# -------------------------------
import os
import zipfile
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import random
from collections import Counter

# -------------------------------
# 3. Upload ZIP if necessary
# -------------------------------
# For Google Colab:
# from google.colab import files
# uploaded = files.upload()

# -------------------------------
# 4. Unzip Dataset
# -------------------------------
zip_path = "asl_alphabet_test.zip"
extract_path = "asl_data"

if os.path.exists(zip_path):
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(extract_path)
    print(f"✅ Unzipped {zip_path} to {extract_path}")
else:
    print(f"⚠️ Zip file {zip_path} not found. Please upload it again.")

# -------------------------------
# 5. Set Parameters
# -------------------------------
IMG_SIZE = 64
DATA_DIR = "asl_data/asl_alphabet_test"

# -------------------------------
# 6. Check Folder Exists
# -------------------------------
if not os.path.exists(DATA_DIR):
    raise FileNotFoundError(f"❌ Folder not found: {DATA_DIR}")

# -------------------------------
# 7. List Images
# -------------------------------
image_files = [f for f in os.listdir(DATA_DIR) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

print(f"✅ Found {len(image_files)} images in {DATA_DIR}")
print("Example files:", image_files[:10])

# -------------------------------
# 8. Load Images and Labels
# -------------------------------
X, y = [], []

for img_name in image_files:
    # ⚠️ FIX THIS IF YOUR FILENAMES ARE LIKE test_A.jpg
    parts = img_name.split('_')
    if len(parts) > 1:
        label = parts[1].split('.')[0].upper()
    else:
        label = parts[0].split('.')[0].upper()

    img_path = os.path.join(DATA_DIR, img_name)
    img = cv2.imread(img_path)

    if img is not None:
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        X.append(img)
        y.append(label)
    else:
        print(f"⚠️ Could not read image: {img_path}")

X = np.array(X) / 255.0

print(f"✅ Loaded {len(X)} images.")
print(f"✅ Labels found: {sorted(set(y))}")
print(f"✅ Label distribution: {Counter(y)}")

if len(X) == 0:
    raise ValueError("❌ No images loaded. Check your folder path and image files.")

# -------------------------------
# 9. Encode Labels
# -------------------------------
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("✅ Classes:", label_encoder.classes_)

# -------------------------------
# 10. Split Data Into Train / Val / Test
# -------------------------------
# Step 1: split off test set
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y_encoded, test_size=0.2, shuffle=True, stratify=y_encoded
)

# Step 2: split temp into train/val
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.25, shuffle=True, stratify=y_temp
)

print(f"✅ Train samples: {len(X_train)}")
print(f"✅ Val samples: {len(X_val)}")
print(f"✅ Test samples: {len(X_test)}")

# -------------------------------
# 11. Build Transfer Learning Model
# -------------------------------
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False  # freeze base model

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128, activation='relu'),
    layers.Dense(len(label_encoder.classes_), activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print(model.summary())

# -------------------------------
# 12. Train Model
# -------------------------------
history = model.fit(
    X_train, y_train,
    epochs=50,
    validation_data=(X_val, y_val)
)

# -------------------------------
# 13. Plot Accuracy
# -------------------------------
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()

# -------------------------------
# 14. Evaluate on Validation Set
# -------------------------------
y_pred_probs_val = model.predict(X_val)
y_pred_classes_val = np.argmax(y_pred_probs_val, axis=1)

all_classes_val = np.unique(np.concatenate([y_val, y_pred_classes_val]))
all_class_names_val = label_encoder.inverse_transform(all_classes_val)

print("\nClassification Report (Validation Set):")
print(classification_report(
    y_val,
    y_pred_classes_val,
    labels=all_classes_val,
    target_names=all_class_names_val,
    zero_division=0
))

# Confusion matrix on validation
cm_val = confusion_matrix(y_val, y_pred_classes_val, labels=all_classes_val)
plt.figure(figsize=(8,6))
sns.heatmap(cm_val, annot=True, fmt='d',
            xticklabels=all_class_names_val,
            yticklabels=all_class_names_val,
            cmap="Blues")
plt.title("Confusion Matrix - Validation Set")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# -------------------------------
# 15. Evaluate on Test Set (unseen data)
# -------------------------------
y_pred_probs_test = model.predict(X_test)
y_pred_classes_test = np.argmax(y_pred_probs_test, axis=1)

all_classes_test = np.unique(np.concatenate([y_test, y_pred_classes_test]))
all_class_names_test = label_encoder.inverse_transform(all_classes_test)

print("\nClassification Report (Test Set):")
print(classification_report(
    y_test,
    y_pred_classes_test,
    labels=all_classes_test,
    target_names=all_class_names_test,
    zero_division=0
))

# Confusion matrix on test set
cm_test = confusion_matrix(y_test, y_pred_classes_test, labels=all_classes_test)
plt.figure(figsize=(8,6))
sns.heatmap(cm_test, annot=True, fmt='d',
            xticklabels=all_class_names_test,
            yticklabels=all_class_names_test,
            cmap="Greens")
plt.title("Confusion Matrix - Test Set")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# -------------------------------
# 16. Show 3 Correct and 3 Incorrect Predictions from Test Set
# -------------------------------
correct_test = np.where(y_test == y_pred_classes_test)[0]
incorrect_test = np.where(y_test != y_pred_classes_test)[0]

print(f"✅ Correct predictions in test set: {len(correct_test)}")
print(f"❌ Incorrect predictions in test set: {len(incorrect_test)}")

# Show 3 correct predictions
if len(correct_test) > 0:
    for idx in random.sample(list(correct_test), min(3, len(correct_test))):
        plt.imshow(X_test[idx])
        true_label = label_encoder.inverse_transform([y_test[idx]])[0]
        pred_label = label_encoder.inverse_transform([y_pred_classes_test[idx]])[0]
        plt.title(f"✅ Correct Prediction: {true_label}")
        plt.axis('off')
        plt.show()
else:
    print("⚠️ No correct predictions in test set to display.")

# Show 3 incorrect predictions
if len(incorrect_test) > 0:
    for idx in random.sample(list(incorrect_test), min(3, len(incorrect_test))):
        plt.imshow(X_test[idx])
        true_label = label_encoder.inverse_transform([y_test[idx]])[0]
        pred_label = label_encoder.inverse_transform([y_pred_classes_test[idx]])[0]
        plt.title(f"❌ Incorrect Prediction:\nTrue={true_label}, Pred={pred_label}")
        plt.axis('off')
        plt.show()
else:
    print("⚠️ No incorrect predictions in test set to display.")
```
